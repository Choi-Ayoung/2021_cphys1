## 1. Pendulum 

```python
import numpy as np
L, g = 9.8, 9.8
dt=0.001
t = np.arange(0,100,dt)
w , theta = np.array([]), np.array([])
w0, theta0 = 1, np.pi

for i in t:
    theta0 = theta0 + dt*w0
    w0 = w0 - dt*(g/L)*np.sin(theta0)
    theta=np.append(theta,theta0)
    w=np.append(w,w0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.xlim(0,50);plt.ylim(0,80)
plt.plot(t,theta);
```

## 3. Damped Pendulum

```python
import numpy as np
beta, L, g =0.1, 9.8, 9.8
dt=0.001
t = np.arange(0,100,dt)
w , theta = np.array([]), np.array([])
w0, theta0 = 1, 1

for i in t:
    theta0 = theta0 + dt*w0
    w0 = w0 - dt*(beta*w0+(g/L)*np.sin(theta0))
    theta=np.append(theta,theta0)
    w=np.append(w,w0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(t,theta);
```

## 4. Forced Pendulum

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
A, v = 1, 1
w=1
state0 = [1.0, 0.0]
t = np.arange(0.0, 100.0, 0.001)

def f(state, t):
    theta, s = state 
    return s, A*np.cos(v*t)-beta*s-(g/L)*np.sin(theta)

states = odeint(f, state0, t)
plt.plot(t,states[:,0]);
```
